### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

display(tienda.head())
display(tienda2.head())
display(tienda3.head())
display(tienda4.head())
```

#1. Análisis de facturación



```python
df = pd.concat([tienda, tienda2, tienda3, tienda4])
df.head()
```

```python
tienda['venta total'] = tienda['Precio'] * tienda['Cantidad de cuotas']
total_ventas_tienda1 = tienda['venta total'].sum()
print(f"El total de ventas de tienda 1 es: {total_ventas_tienda1}")

tienda2['venta total'] = tienda2['Precio'] * tienda2['Cantidad de cuotas']
total_ventas_tienda2 = tienda2['venta total'].sum()
print(f"El total de ventas de tienda 2 es: {total_ventas_tienda2}")

tienda3['venta total'] = tienda3['Precio'] * tienda3['Cantidad de cuotas']
total_ventas_tienda3 = tienda3['venta total'].sum()
print(f"El total de ventas de tienda 3 es: {total_ventas_tienda3}")

tienda4['venta total'] = tienda4['Precio'] * tienda4['Cantidad de cuotas']
total_ventas_tienda4 = tienda4['venta total'].sum()
print(f"El total de ventas de tienda 4 es: {total_ventas_tienda4}")

```

```python

df['venta total'] = df['Precio'] * df['Cantidad de cuotas']
```

```python
total_ventas_todas_tiendas = df['Venta Total'].sum()
print(f"La sumatoria del total de ventas de todas las tiendas es: {total_ventas_todas_tiendas}")
```

```python
import matplotlib.pyplot as plt

ventas_por_tienda = {
    'Tienda 1': total_ventas_tienda1,
    'Tienda 2': total_ventas_tienda2,
    'Tienda 3': total_ventas_tienda3,
    'Tienda 4': total_ventas_tienda4
}

nombres = ventas_por_tienda.keys()
ventas = ventas_por_tienda.values()

plt.figure(figsize=(8, 8))
plt.pie(ventas, labels=nombres, autopct='%1.1f%%', startangle=140)
plt.title('Porcentaje de Ventas por Tienda')
plt.axis('equal')
plt.show()
```

# 2. Ventas por categoría

```python
pd.options.display.float_format = '{:.2f}'.format
ventas_por_categoria = df.groupby('Categoría del Producto')['venta total'].sum().sort_values(ascending=False)
print("Total de ventas por categoría:")
display(ventas_por_categoria)
```

```python
import matplotlib.pyplot as plt

category_sales = ventas_por_categoria

category_sales_sorted = category_sales.sort_values(ascending=False)

plt.figure(figsize=(12, 6))
category_sales_sorted.plot(kind='bar')
plt.title('Ventas por Categoría (Ordenadas por Monto)')
plt.xlabel('Categoría del Producto')
plt.ylabel('Venta Total')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python

transacciones_por_producto = df.groupby('Producto').size()


productos_mas_transacciones = transacciones_por_producto.sort_values(ascending=False)
print("Productos con más transacciones (basado en Fecha de Compra):")
display(productos_mas_transacciones.head())


productos_menos_transacciones = transacciones_por_producto.sort_values(ascending=True)
print("\nProductos con menos transacciones (basado en Fecha de Compra):")
display(productos_menos_transacciones.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


top_products = productos_mas_transacciones.head(10).index
top_products_df = df[df['Producto'].isin(top_products)][['Producto', 'Categoría del Producto']].drop_duplicates()
top_products_df = top_products_df.set_index('Producto').loc[top_products].reset_index() # Maintain order
top_products_df['Cantidad de Transacciones'] = productos_mas_transacciones.head(10).values

plt.figure(figsize=(14, 7))
sns.barplot(x='Producto', y='Cantidad de Transacciones', hue='Categoría del Producto', data=top_products_df, dodge=False)
plt.title('Top 10 Productos con Más Transacciones por Categoría')
plt.xlabel('Producto')
plt.ylabel('Número de Transacciones')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


bottom_products = productos_menos_transacciones.head(10).index
bottom_products_df = df[df['Producto'].isin(bottom_products)][['Producto', 'Categoría del Producto']].drop_duplicates()
bottom_products_df = bottom_products_df.set_index('Producto').loc[bottom_products].reset_index() # Maintain order
bottom_products_df['Cantidad de Transacciones'] = productos_menos_transacciones.head(10).values


plt.figure(figsize=(14, 7))
sns.barplot(x='Producto', y='Cantidad de Transacciones', hue='Categoría del Producto', data=bottom_products_df, dodge=False)
plt.title('Top 10 Productos con Menos Transacciones por Categoría')
plt.xlabel('Producto')
plt.ylabel('Número de Transacciones')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## **6. Desarrollo final consejos para Sr. Juan**

*"Determinación de la tienda menos eficiente"*


Para identificar cuál de las cuatro tiendas presenta el peor desempeño, se analizaron cinco dimensiones clave: ingresos totales, categorías más vendidas, calificación promedio, volumen de productos vendidos y costo de envío promedio. Cada indicador permite observar distintos aspectos de la eficiencia general de las operaciones de cada tienda.

*- 1. Ingresos totales*

Los ingresos totales representan el valor económico generado por cada tienda. Según los datos obtenidos:

Tienda 1: 3.422.717.700
Tienda 2: 3.337.189.200
Tienda 3: 3.243.816.900
Tienda 4: 3.098.827.200

La tienda con menor ingreso es la **Tienda 4**, lo cual indica que aporta el menor retorno económico dentro del sistema. Este es un indicador fuerte de bajo desempeño comercial.

*- 2. Categorías de productos más y menos vendidas*

Si bien los datos de categorías se presentan de manera global y no diferenciados por tienda, pueden relacionarse indirectamente con el desempeño. El total por categoría muestra que los artículos con mayor facturación provienen de Electrónicos y Electrodomésticos, mientras que las categorías con menor volumen económico corresponden a Juguetes, Artículos para el hogar y Libros.

Si la Tienda 4 posee mayor proporción de ventas en categorías de bajo valor o menor volumen en categorías de alto rendimiento, esto contribuye a su menor ingreso total.

*- 3. Calificación promedio*

Las calificaciones promedio muestran la percepción de los clientes sobre la experiencia de compra:

* Tienda 1: 3.98
* Tienda 2: 4.04
* Tienda 3: 4.05
* Tienda 4: 4.00

Aunque la diferencia no es grande, la Tienda 4 se encuentra nuevamente entre las dos con peor evaluación. Esto sugiere satisfacción ligeramente inferior y posible impacto en la retención de clientes.

*- 4. Productos más y menos vendidos*

Los productos con mayor número de transacciones incluyen mesas de noche, microondas y juguetes específicos, mientras que los menos vendidos incluyen libros, mochilas y ciertos electrónicos.

Si la Tienda 4 registra menor frecuencia de venta en los productos de mayor rotación o se apoya fuertemente en productos de baja demanda, esto contribuye a su menor rendimiento.

*- 5. Costo de envío promedio*

El costo de envío promedio por tienda es:

* Tienda 1: 26.018
* Tienda 2: 25.216
* Tienda 3: 24.805
* Tienda 4: 23.459

La Tienda 4 posee el costo de envío más bajo. Este indicador, aunque positivo, no es suficiente para compensar su bajo nivel de ingresos y su calificación relativamente menor. De hecho, el bajo costo de envío puede sugerir menor complejidad logística, posiblemente asociada a un volumen de pedidos más reducido.

*- Conclusión*

Luego de analizar todos los factores disponibles, se determina que la Tienda 4 es la menos eficiente. Esto se evidencia principalmente en:

- Su menor ingreso total, lo cual refleja menor capacidad de generar ventas o menor valor por operación.
- Una calificación promedio inferior en comparación con las otras tiendas.
- La posibilidad de que maneje menos volumen de productos de alto rendimiento, lo que se alinea con su menor facturación.
- Un costo de envío más bajo, que podría estar asociado a menos actividad logística.

En conjunto, estos elementos sostienen que la Tienda 4 presenta el rendimiento más débil y, por lo tanto, es la candidata más razonable para ser vendida.
